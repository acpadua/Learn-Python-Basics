# What is a function?

It is a piece of reusable code, that aims at solving a particular task.
So, you can call a function instead of writting code by yourself.

Examples:

type() - returns type of a value

Suppose you have a list with the age of students in your class.
You can get the maximum of that list using the function max()

```python
student_age = [20,25,24,27,37,28,32]
max(student_age)
```

We can assign a variable to the oldest student

```python
oldest = max(student_age)
print(oldest)
```

```python
# Another exaple of a built-in function

round(5.879, 2)

# This function takes 2 inputs:
# input 1: value
# input 2: decimal values to round

#Note: inputs are also called arguments in python
# The result of that the function produces is called an output
```

```python
# If you don't specify the input2, python assumes to round to number to the closest integer.
round(5.879)
```

# What is a method?

A method is a function that belongs to an object.
Everyhting in python is an object, but there are different types of objects.

Examples of objects:
* Strings
* Integers / Floats
* Boolean
* Lists

### List methods: examples

Let's see an example of a method associated to a list object.

```python
student_age.index(24) # gives the index of the element "24" in the list
student_age.count(32) # counts number of times that en elements occurs in a list
```

```python
student_age.append(26)
student_age # check it is updated
```

# Packages

Python has so many functions and methods that the number of scrips becomes messy!

For managing this, the concept of packaging was created.

You can think of packages as a directoty of python scrips. Each script is a module, which specifies funtions, methods and new python types aimed at solving particular problems.

From the thousands of packages available, some of the most wisely used in data science:

* NumPy: to offically work with arrays
* Matplotlib: for data visualization
* scikit-learn: for machine learning
