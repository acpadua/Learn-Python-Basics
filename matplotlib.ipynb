Create a line plot

```python
#plt.plot(date, weight)
#plt.show()
#plt.clf() # cleans it up again so you can start afresh

# Customization
plt.plot(date, weight)

#Strings
xlab = 'Date [year/month/day]'
ylab = 'Weight [in Kg]'
title = 'Weight variation'

# Add axis labels
plt.xlabel(xlab)
plt.ylabel(ylab)

# Add title
plt.title(title)

# Define axis limits
plt.ylim(40, 55)

# Definition of tick_val and tick_lab of y axis
y_tick_val = [40, 45, 50, 55]
y_tick_lab = ['40Kg', '45Kg', '50Kg', '55Kg']

# Define ticks of the x axis
plt.xticks([0,1,2], date)

# Adapt the ticks on the y-axis
plt.yticks(y_tick_val, y_tick_lab)

plt.plot(date, weight)

# Show plot
plt.show()
```
